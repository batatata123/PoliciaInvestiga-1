```
13) Contagens
numeros = []

for i in range(20):
    numeros.append(int(input("Digite um número: ")))

pares = 0
impares = 0
maior50 = 0
menor7 = 0

for n in numeros:
    if n % 2 == 0:
        pares += 1
    else:
        impares += 1

    if n > 50:
        maior50 += 1
    if n < 7:
        menor7 += 1

print(f"Pares: {pares}")
print(f"Ímpares: {impares}")
print(f"Maiores que 50: {maior50}")
print(f"Menores que 7: {menor7}")
```

```
12) Soma de Vetores
v1 = []
v2 = []

for i in range(10):
    v1.append(int(input(f"Vetor 1 (posição {i}): ")))

for i in range(10):
    v2.append(int(input(f"Vetor 2 (posição {i}): ")))

resultado = []
for i in range(10):
    resultado.append(v1[i] + v2[i])

print("Resultado da soma:", resultado)
```

```
11) Buscar e Mostrar Posição
numeros = []

for i in range(8):
    numeros.append(int(input("Digite um número: ")))

busca = int(input("Digite um número para buscar a posição: "))

if busca in numeros:
    print(f"Encontrado na posição: {numeros.index(busca)}")
else:
    print("Não encontrado")
```

```
5) Transposta
matriz = [[1, 2, 3], [4, 5, 6]]
transposta = []

for i in range(len(matriz[0])):
    linha = []
    for j in range(len(matriz)):
        linha.append(matriz[j][i])
    transposta.append(linha)

for linha in transposta:
    print(linha)
```

```
4) Diagonal Principal
matriz = [[1, 2], [3, 4]]

for i in range(len(matriz)):
    print(matriz[i][i])
```

```
3) Soma dos Elementos
matriz = [[1, 2], [3, 4]]
soma = 0

for linha in matriz:
    for valor in linha:
        soma += valor

print(soma)
```

```
2) Multiplicação por Número
matriz = [[1, 2], [3, 4]]
numero = 2

for i in range(2):
    for j in range(2):
        matriz[i][j] *= numero

for linha in matriz:
    print(linha)
```

```
1) Soma de Matrizes
matriz1 = [[1, 2], [3, 4]]
matriz2 = [[5, 6], [7, 8]]
resultado = [[0, 0], [0, 0]]

for i in range(2):
    for j in range(2):
        resultado[i][j] = matriz1[i][j] + matriz2[i][j]

for linha in resultado:
    print(linha)
```
